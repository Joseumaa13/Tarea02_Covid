```python
table1 = go.Figure(data=[go.Table(
    header=dict(values=list(table.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[general.hom_posi, general.muj_posi, general.adul_posi, general.am_posi, general.nue_posi],
               fill_color='lavender',
               align='left'))
])

table1.show()
```
